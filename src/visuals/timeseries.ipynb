```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Parameter
seed = 42
iter = 1e+06
alpha = 1e-03
lr = 1e-03
tol = 1e-06

file_path = f'/home/mori-lab/shimizu/ACO/ACO-on-SpinGlass/src/sampling/grad_method/data/seed{seed}/iter{iter:.0e}/timeseries_alpha{alpha:.0e}_lr{lr:.0e}_tol{tol:.0e}.csv'

data = pd.read_csv(file_path)
data_alpha = data['alpha'] * -800
data_energy = data['energy']

plt.plot(data_alpha)
plt.plot(data_energy)
plt.xscale('log')
plt.xlim(1e3, iter)
plt.show()
print(int(np.sum(data_alpha) / -800))
```
